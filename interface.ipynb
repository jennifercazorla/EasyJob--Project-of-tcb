```python
"""
interface
"""

import tkinter as tk
from tkinter import *
from functools import partial
from easyjob_ import easyjob
from oferta import oferta
from empresa import empresa
from empleado import empleado

# definimos ventana principal

class principal(object):
    def __init__(self):
        self.dic_easyjob=easyjob() #creación de instancia para relacionar nuestra clase principal con la interfaz gráfica
        self.empresa = empresa()
        self.empleado = empleado()
        # Creamos nuestra ventana principal
        self.vp = tk.Tk()
        
        # Esta ventana principal va a tener las dimensiones:
        self.vp.geometry('900x600')
        # En este caso nuestro fondo va a ser de un tono beige
        self.vp.configure(bg= 'beige')
        # El título de esta ventana va a ser el nombre de nuestra web
        self.vp.title('EASYJOB')
        # Creamos los dos botones principales de mi ventana, y cada boton va a tener su comando correspondiente. 
        # Encontramos las funciones de los command en este mismo fichero. Ambas funciones abrirán nuevas ventanas para tratar campos diferentes. 
        self.b_empleado = tk.Button(text = "Empleado", font=('Gotham',20), width=20,fg='blue',
                                 bg='black',command= self.empleado_sesion).place(x=100,y=280)
        self.b_empresa = tk.Button(text = "Empresa",font=('Gotham',20), width=20,fg='blue',
                                bg='black',command= self.empresa_sesion).place(x=500,y=280)
        
        self.salir = tk.Button(text = "Exit",font=('Gotham',20), width=20,fg='white',bg='red',
                                 command= self.vp.destroy).place(x=300,y=550)
        
        self.vp.mainloop()
        
    """
    Parte del trabajador 
    """
    
    def empleado_sesion(self):
        
        #Creamos la nueva ventana correspondiente al trabajador
        v_empleado = tk.Toplevel(self.vp)
        # Esta ventana tendrá las siguientes dimensiones: 
        v_empleado.geometry("275x150")
        
        v_empleado.title("Apartado para usuarios que buscan trabajo")
        #Apartado para poner el usuario e iniciar sesión
        etiqueta_usuario = tk.Label(v_empleado, text= "Usuario:").grid(column=0, row=1)
        usuario1 = tk.StringVar()
        usuario1.set("")
        entry_nombre = tk.Entry(v_empleado, textvariable=usuario1).grid(column=1, row=1)
        etiqueta_contraseña = tk.Label(v_empleado, text= "Contraseña:").grid(column=0, row=2)
        contraseña1 = tk.StringVar()
        contraseña1.set("")
        entry_contraseña = tk.Entry(v_empleado, textvariable=contraseña1).grid(column=1, row=2)
      
        #Comprobar el usuario, indica si no existe. La función empleada también se encuentra en esta clase
        verifica_empleado_parametros = partial(self.usuario_parametros, v_empleado, usuario1)

        abrir_registro = partial(self.crear_empleado)
      
        Iniciar=tk.Button(v_empleado,text="Iniciar sesión", font=('Gotham',10), 
                          width=20,fg='white',bg='forest green',
                          command=verifica_empleado_parametros).grid(column=1,row=3)
       
        Cancelar=tk.Button(v_empleado,text="Cancelar",font=('Gotham',10), width=20,fg='white',bg='red',
                                command= v_empleado.destroy).grid(column=1,row=5)
       
        Registrarse = tk.Button(v_empleado, text = "Registrarse",font=('Gotham',10), width=20,fg='white',bg='forest green',
                                command = abrir_registro).grid(column=1,row=4)
        # Es para que nuestra ventana este sobre su maestra
        v_empleado.transient()

       #Para evitar interactuar con otra ventana
        v_empleado.grab_set()

       # Wait for the window to end 
        self.vp.wait_window(v_empleado)
        
        if usuario1.get():
          existe = self.dic_easyjob.comprueba_usuario_empleado(usuario1.get(), contraseña1.get())
          print(existe)
          
          """
          Que imprima la info
          """
          
          if existe:
              v_sesion = tk.Toplevel(self.vp)
              # Tiene las siguientes dimensiones
              v_sesion.geometry("400x175")
              # El fondo de la ventana es un verda clarito
              v_sesion.configure(bg= 'pale green')
              v_sesion.title("ESPACIO EMPLEADO") 
              #Creación de la barra para el menú
              barramenu = tk.Menu(v_sesion)
              verdatosaux=partial(self.verdatosempleado, usuario1.get())       
       #Items
              MenuEmpleado = tk.Menu(barramenu)
   
       #Comandos del menú. Estos corresponden a funciones descritas en este mismo fichero. 
              MenuEmpleado.add_command(label="Aplicar a ofertas", command=v_sesion.destroy)
              MenuEmpleado.add_command(label="Ver datos", command=verdatosaux)
              MenuEmpleado.add_command(label="Salir", command = v_sesion.destroy)
       
       #Etiquetamos el menú: 
              barramenu.add_cascade(label="Empleado", menu = MenuEmpleado)
              v_sesion.config(menu = barramenu)
              
              
              aplicarofertasaux=partial(self.aplicar_oferta, empresa)

       
       #Botones de nuestra ventana, van a hacer la misma función que el menú
              veroferta = tk.Button(v_sesion, text = "Aplicar a ofertas", font=('Gotham',20), width=20,fg='blue',
                                bg='black', command=aplicarofertasaux).pack(side=TOP, fill=BOTH, expand=True)
              verdatos = tk.Button(v_sesion, text = "Ver mis datos", font=('Gotham',20), width=20,fg='blue',
                                bg='black', command= verdatosaux).pack(side=TOP, fill=BOTH, expand=True)

              salir = tk.Button(v_sesion, text = "Exit",font=('Gotham',20), width=20,fg='white',bg='red',
                                command= v_sesion.destroy).pack(side=TOP, fill=BOTH, expand=True)
       

              v_sesion.mainloop()
              
              v_sesion.transient()
              
              v_sesion.grab_set()
              
              self.vp.wait_window(v_sesion)
   
          
        else:
              tk.messagebox.showinfo(title='Error', message='No existe el usuario')
              
             
    """
    Este funcion es la encargada de ver si el  campo introducido estña vació
    """
    
    def usuario_parametros(self, ventana,usuario):
        """
        Auxiliar function to be able to send messageboxes
        """ 
        # Mirar si algun esta empty
        if not all([usuario.get()]):
            
            tk.messagebox.showinfo(title='Error', message='El campo está vacío')
        else:
            ventana.destroy()
    
        
        """
        empleado_sesion se encarga de abrir una nueva ventana tras iniciar sesión. En caso de estar 
        registrado, el programa mostrará la información contenida en el perfil
        """

#La siguiente funcion nos muestra los datos de un empleado
    def verdatosempleado (self, usuario):
       existe = self.dic_easyjob.muestra_empleado(usuario)
       voo = tk.Toplevel(self.vp)
       voo.geometry("300x350")
       voo.title("Datos usuario") 
                
       # Nombre usuario
       etiqueta_usuario = tk.Label(voo, text= "Nombre usuario:").grid(column=0, row=10)
       v_usuario = tk.StringVar()
       v_usuario.set(existe.usuario)
       entry_usuario = tk.Entry(voo, textvariable=v_usuario, state=tk.DISABLED).grid(column=1, row=10)

       # Nombre
       etiqueta_nombre = tk.Label(voo, text= "Nombre:").grid(column=0, row=1)
       v_nombre = tk.StringVar()
       v_nombre.set(existe.nombre)
       entry_nombre = tk.Entry(voo, textvariable=v_nombre,state=tk.DISABLED).grid(column=1, row=1)
         
       # Apellidos
       etiqueta_apellidos = tk.Label(voo, text= "Apellidos:").grid(column=0, row=2)
       v_apellidos = tk.StringVar()
       v_apellidos.set(existe.apellidos)
       entry_apellidos = tk.Entry(voo, textvariable=v_apellidos,state=tk.DISABLED).grid(column=1, row=2)
           
                 # Ciudad de interes
       etiqueta_lugar = tk.Label(voo, text= "Ciudad a trabajar:").grid(column=0, row=3)
       v_lugar = tk.StringVar()
       v_lugar.set(existe.lugar)
       entry_lugar = tk.Entry(voo, textvariable=v_lugar,state=tk.DISABLED).grid(column=1, row=3)
         
                 # Edad
       etiqueta_edad = tk.Label(voo, text= "Fecha de nacimiento:").grid(column=0, row=4)
       v_edad = tk.StringVar()
       v_edad.set(existe.edad)
       e_edad = tk.Entry(voo, textvariable=v_edad,state=tk.DISABLED).grid(column=1, row=4)
         
                 # Nivel de estudios: en este caso es un spinbox, tienes que elegir el que cuadre con lo que has hecho
       etiqueta_nivel = tk.Label(voo, text= "Nivel de Estudios:").grid(column=0, row=5)
       v_nivel = tk.StringVar()
       v_nivel.set(existe.nivel)
       e_nivel = tk.Entry(voo, textvariable=v_nivel,state=tk.DISABLED).grid(column=1, row=5)
             
                
                 # Especificar estudio
                 
       etiqueta_estudio = tk.Label(voo, text= "Especificar estudio:").grid(column=0, row=6)
       v_estudio = tk.StringVar()
       v_estudio.set(existe.estudio)
       entry_dir = tk.Entry(voo, textvariable=v_estudio,state=tk.DISABLED).grid(column=1, row=6)
                 
                 # Movil
                 
       etiqueta_mv = tk.Label(voo, text= "Nº de movil").grid(column=0, row=7)
       v_mv = tk.StringVar()
       v_mv.set(existe.movil)
       etry_mv = tk.Entry(voo, textvariable=v_mv,state=tk.DISABLED).grid(column=1, row=7)
                        
                 # email
         
       etiqueta_email = tk.Label(voo, text= "Email:").grid(column=0, row=8)
       v_email = tk.StringVar()
       v_email.set(existe.email)
       e_email = tk.Entry(voo, textvariable=v_email,state=tk.DISABLED).grid(column=1, row=8)
                 
                 #  Jornada: es un spinbox ya que solo hay dos resouestas validas
       etiqueta_tiempo = tk.Label(voo, text= "Tipo de jornada:").grid(column=0, row=9)
       v_tiempo = tk.StringVar()
       v_tiempo.set(existe.tiempo)
       entry_tiempo = tk.Entry(voo, textvariable=v_tiempo,state=tk.DISABLED).grid(column=1, row=9) 
            
                 #Contraseña
       etiqueta_contraseña = tk.Label(voo, text= "Contraseña:").grid(column=0, row=11)
       v_contraseña = tk.StringVar()
       v_contraseña.set(existe.contraseña)
       entry_contraseña = tk.Entry(voo, textvariable=v_contraseña, state=tk.DISABLED).grid(column=1, row=11)
                 

       exitboton=Button(voo,text="Salir",font=('Gotham',10), width=20,fg='white',bg='red',
                                          command= voo.destroy).grid(column=1,row=12)
         
         
                 # Esta ventana este sobre su maestra
       voo.transient()
         
                 #Para evitar interactuar con otra ventana
       voo.grab_set()
         
                 # Wait for the window to end 
       self.vp.wait_window(voo)
            

#En caso que no exista ya el usuario, esta funcion te permite guardar la informacion 
#de un usuario nuevo
    def crear_empleado(self):
        #Ventana para poder crearse un perfil        
        v_login = tk.Toplevel(self.vp)
        v_login.geometry("350x100")
        v_login.title("Añadir perfil")    
        
        # Etiqueta que indica que tiene que hacer
        etiqueta_incio= tk.Label(v_login, text= "Rellenar con tus datos personales:")
        etiqueta_incio.grid(column=0, row=0)
                
        etiqueta_usuario1=tk.Label(v_login,text="usuario: ",width=10)
        etiqueta_usuario1.grid(column=0,row=1)
        usuario1=tk.StringVar()
        usuario1.set("")
        usuario=tk.Entry(v_login,textvariable=usuario1,width=15)
        usuario.grid(column=1,row=1)
        etiqueta_contraseña1=tk.Label(v_login,text="contraseña: ",width=10)
        etiqueta_contraseña1.grid(column=0,row=2)
        contraseña1=tk.StringVar()
        contraseña1.set("")
        contraseña=tk.Entry(v_login,textvariable=contraseña1,width=15)
        contraseña.grid(column=1,row=2)        
        # de la llibreria functools
        # assignar parcial per a funció, per a poder assignar directament command amb variables
        
        verifica_usuario_params=partial(self.usuario_parametros, v_login, usuario1)

        # Programar botó
        btnAsignar=tk.Button(v_login,text="Siguiente",  command = verifica_usuario_params).grid(column=0,row=3)
        btnSortir=tk.Button(v_login,text="Cancelar", command = v_login.destroy).grid(column=1,row=3)
        v_login.transient()
        v_login.grab_set()
        self.vp.wait_window(v_login)
        
        """
        Comprobar que el usuario no existe
        """
        if usuario1.get():
           existe=self.dic_easyjob.comprueba_usuario_empleado(usuario1.get(), contraseña1.get())
           #print(existe)
           if not existe:
               v_ingreso = tk.Toplevel(self.vp)
               v_ingreso.geometry("300x350")
               v_ingreso.title("Alta de usuario") 
               etiqueta_nombre = tk.Label(v_ingreso, text= "Nombre:").grid(column=0, row=1)
               v_nombre = tk.StringVar()
               v_nombre.set("")
               entry_nombre = tk.Entry(v_ingreso, textvariable=v_nombre).grid(column=1, row=1)
        
                # Nombre y apellidos

               etiqueta_apellidos = tk.Label(v_ingreso, text= "Apellidos:").grid(column=0, row=2)
               v_apellidos = tk.StringVar()
               v_apellidos.set("")
               entry_apellidos = tk.Entry(v_ingreso, textvariable=v_apellidos).grid(column=1, row=2)
          
                # Ciudad de interes
               etiqueta_lugar = tk.Label(v_ingreso, text= "Ciudad a trabajar:").grid(column=0, row=3)
               v_lugar = tk.StringVar()
               v_lugar.set("")
               entry_lugar = tk.Entry(v_ingreso, textvariable=v_lugar).grid(column=1, row=3)
        
                # Edad
               etiqueta_edad = tk.Label(v_ingreso, text= "Fecha de nacimiento:").grid(column=0, row=4)
               v_edad = tk.StringVar()
               v_edad.set("")
               e_edad = tk.Entry(v_ingreso, textvariable=v_edad).grid(column=1, row=4)
        
               # Nivel de estudios: en este caso es un spinbox, tienes que elegir el que cuadre con lo que has hecho
               #etiqueta_nivel = tk.Label(v_ingreso, text= "Nivel de Estudios:").grid(column=0, row=5)
               #v_nivel = tk.Spinbox(v_ingreso, values=('ESO','Bachillerato','Ciclo medio','Ciclo superior','Universidad','Master','Doctorado')).grid(column=1, row=5)
               etiqueta_nivel = tk.Label(v_ingreso, text= "Nivel:").grid(column=0, row=5)
               v_nivel = tk.StringVar()
               v_nivel.set("")
               e_nivel = tk.Entry(v_ingreso, textvariable=v_nivel).grid(column=1, row=5)
                   
                # Especificar estudio
                
               etiqueta_estudio = tk.Label(v_ingreso, text= "Especificar estudio:").grid(column=0, row=6)
               v_estudio = tk.StringVar()
               v_estudio.set("")
               etry_dir = tk.Entry(v_ingreso, textvariable=v_estudio).grid(column=1, row=6)
                
                # Movil
                
               etiqueta_mv = tk.Label(v_ingreso, text= "Nº de movil").grid(column=0, row=7)
               v_mv = tk.StringVar()
               v_mv.set("")
               etry_mv = tk.Entry(v_ingreso, textvariable=v_mv).grid(column=1, row=7)
                       
                # email
        
               etiqueta_email = tk.Label(v_ingreso, text= "Email:").grid(column=0, row=8)
               v_email = tk.StringVar()
               v_email.set("")
               e_email = tk.Entry(v_ingreso, textvariable=v_email).grid(column=1, row=8)
                
                #  Jornada: es un spinbox ya que solo hay dos resouestas validas
              # etiqueta_tiempo = tk.Label(v_ingreso, text= "Tipo de jornada:").grid(column=0, row=9)
              # v_tiempo = tk.Spinbox(v_ingreso, values=('Jornada completa','Jornada media')).grid(column=1, row=9) 
               etiqueta_tiempo = tk.Label(v_ingreso, text= "Jornada:").grid(column=0, row=9)
               v_tiempo = tk.StringVar()
               v_tiempo.set("")
               e_tiempo = tk.Entry(v_ingreso, textvariable=v_tiempo).grid(column=1, row=9)
                               
                # Nombre usuario
               etiqueta_usuario = tk.Label(v_ingreso, text= "Nombre usuario:").grid(column=0, row=10)
               v_usuario = tk.StringVar()
               v_usuario.set(usuario1.get())
               entry_usuario = tk.Entry(v_ingreso, textvariable=v_usuario, state=tk.DISABLED).grid(column=1, row=10)
                 
                #Contraseña
               etiqueta_contraseña = tk.Label(v_ingreso, text= "Contraseña:").grid(column=0, row=11)
               v_contraseña = tk.StringVar()
               v_contraseña.set(contraseña1.get())
               entry_contraseña = tk.Entry(v_ingreso, textvariable=v_contraseña, state = tk.DISABLED).grid(column=1, row=11)
                
                
               # Creación del metodo para guardar los datos
               crear_empleado_param=partial(self.crear_empleado_aux,v_ingreso, v_nombre, v_apellidos, v_lugar, v_edad,v_nivel, v_estudio, v_mv, v_email,v_tiempo, v_usuario, v_contraseña)
                
                
               # Botón para guardar y una para salir (todavía no sabemos como guardar la función añadida, por ahora ambos botones son para salir)
               Aceptar=tk.Button(v_ingreso,text="Aceptar", font=('Gotham',10), width=15,fg='white',bg='forest green',
                                       command= crear_empleado_param)
               Aceptar.grid(column=1,row=12)
               Cancelar=tk.Button(v_ingreso,text="Salir",font=('Gotham',10), width=15,fg='white',bg='red',
                                         command= v_ingreso.destroy)
               Cancelar.grid(column=1,row=13)
        
        
                # Esta ventana este sobre su maestra
               v_ingreso.transient()
        
                #Para evitar interactuar con otra ventana
               v_ingreso.grab_set()
        
                # Wait for the window to end 
               self.vp.wait_window(v_ingreso)
          
            
           else: 
                tk.messagebox.showinfo(title='Error', message='Ya existe una cuenta, por favor intenta con otro nombre de usuario')
  
                
    def crear_empleado_aux(self, ventana, nombre, apellidos, lugar, edad,nivel, estudio, mv, email,tiempo, usuario, contraseña):
    # Verificar que los campos requeridos no estén vacíos
        if nombre.get() == "" or apellidos.get() == "" or lugar.get() == "" or edad.get() == "" or nivel.get() == "" or estudio.get() == "" or mv.get() == "" or email.get() == "" or tiempo.get() == "" or usuario.get() == "" or contraseña.get() == "":
           tk.messagebox.showinfo(title='Error', message='Falta por rellenar algún hueco')
        else:
            self.dic_easyjob.alta_empleado(nombre.get(), apellidos.get(), lugar.get(), edad.get(), nivel.get(), estudio.get(), mv.get(), email.get(), tiempo.get(), usuario.get(), contraseña.get())
            tk.messagebox.showinfo(title='Añadido', message='Añadido el empleado')
            ventana.destroy()
            
    #Ventana que imprime la info de una oferta para poder darle a aceptar y rechazar         
    def aplicar_oferta(self, empresa):
         ofertas = self.empresa.ver_ofertas()
         print(ofertas)
         
         
         for oferta in ofertas:
             nombre, lugar, tarea, salario, jornada, vacantes, requisitos = oferta
             nombre = oferta[0]
             lugar = oferta[1]
             tarea = oferta[2]
             salario = oferta[3]
             jornada = oferta[4]
             vacantes = oferta[5]
             requisitos = oferta[6]
        #Ventana para poder ver una oferta
             ap_oferta = tk.Toplevel(self.vp)
        #Dimensiones de la ventana
             ap_oferta.geometry("600x350")
             ap_oferta.title("Información de la oferta")    

            #Etiqueta que indica que tienes que hacer
             etiqueta_intro= tk.Label(ap_oferta, text= "Rellena la información para tu anuncio:")
             etiqueta_intro.grid(column=0, row=0)
    
            # Titulo
             etiqueta_titulo = tk.Label(ap_oferta, text= "Oferta:").grid(column=0, row=1)
             v_titulo = tk.StringVar()
             v_titulo.set(nombre)
             entry_titulo = tk.Entry(ap_oferta, textvariable=v_titulo, state=tk.DISABLED)
             entry_titulo.grid(column=1, row=1)
            
            # Lugar:
             etiqueta_lugar = tk.Label(ap_oferta, text= "Lugar:").grid(column=0, row=2)
             v_lugar = tk.StringVar()
             v_lugar.set(lugar)
             entry_lugar = tk.Entry(ap_oferta, textvariable=v_lugar, state=tk.DISABLED)
             entry_lugar.grid(column=1, row=2)
      
                # Tarea/posición 
             etiqueta_tarea = tk.Label(ap_oferta, text= "Tarea a realizar:").grid(column=0, row=3)
             v_tarea = tk.StringVar()
             v_tarea.set(tarea)
             entry_tarea = tk.Entry(ap_oferta, textvariable=v_tarea, state=tk.DISABLED)
             entry_tarea.grid(column=1, row=3)
        
                # Salario
             etiqueta_salario = tk.Label(ap_oferta, text= "Salario:").grid(column=0, row=4)
             v_salario = tk.StringVar()
             v_salario.set(salario)
             entry_salario = tk.Entry(ap_oferta, textvariable=v_salario, state=tk.DISABLED)
             entry_salario.grid(column=1, row=4)
        
                # Jornada
             etiqueta_jornada = tk.Label(ap_oferta, text= "Jornada:").grid(column=0, row=5)
             v_jornada = tk.StringVar()
             v_jornada.set(jornada)
             entry_jornada = tk.Entry(ap_oferta, textvariable=v_jornada, state=tk.DISABLED)
             entry_jornada.grid(column=1, row=5)        
                
                # Vacantes
             etiqueta_vacante = tk.Label(ap_oferta, text= "Número de vacantes:").grid(column=0, row=6)
             v_vacante = tk.StringVar()
             v_vacante.set(vacantes)
             entry_vacante = tk.Entry(ap_oferta, textvariable=v_vacante, state=tk.DISABLED)
             entry_vacante.grid(column=1, row=6)
                
                # Requisitos
             etiqueta_requistos = tk.Label(ap_oferta, text= "Introduce los requisios básicos que esperas en un trabajador:").grid(column=0, row=7)
             v_requisitos = tk.StringVar()
             v_requisitos.set(requisitos)
             entry_requisitos = tk.Entry(ap_oferta, textvariable=v_requisitos, state=tk.DISABLED)
             entry_requisitos.grid(column=1, row=7)
                
             crear_aceptar_param=partial(self.crear_lista_aceptadas_aux, ap_oferta)
             
             borarr_oferta_param=partial(self.borrar_oferta_aux, ap_oferta )#, v_titulo, v_lugar, v_tarea, v_salario, v_jornada,v_vacante, v_requisitos)
              
                # Botón para salir 
             Aceptar=tk.Button(ap_oferta,text="Aplicar",font=('Gotham',10), width=20,fg='white',bg='red',
                                         command= crear_aceptar_param).grid(column=0,row=8)
                
             Cancelar=tk.Button(ap_oferta,text="No me interesa",font=('Gotham',10), width=20,fg='white',bg='red',
                                         command= borarr_oferta_param).grid(column=1,row=8)
                
                # Esta ventana este sobre su maestra
             ap_oferta.transient()
        
                #Para evitar interactuar con otra ventana
             ap_oferta.grab_set()
        
                # Wait for the window to end 
             self.vp.wait_window(ap_oferta)
        
             ap_oferta.mainloop()
    
    #La siguiente funcion se encargaba de poder crear una lista con todas las ofertas 
    #añadidas pero no lo hemos terminado de desarrollar, por lo que solo envia un 
    #mensaje de que esta hecho
    
    def crear_lista_aceptadas_aux(self, ap_oferta):
        
        #No realiaza ninguna funciona, solo estetico
        tk.messagebox.showinfo(title='Añadida', message='Oferta aceptada!')    
        
        ap_oferta.destroy()
        
     #La siguiente funcion se encargaba de poder eliminar las ofertas que no te interesabn 
     # y que no volviesen a salir pero no lo hemos terminado de desarrollar, por lo que solo envia un 
     #mensaje de que esta hecho
    def borrar_oferta_aux(self, ventana): #, v_titulo, v_lugar, v_tarea, v_salario, v_jornada,v_vacante, v_requisitos):
        
            #oft = oferta(v_titulo.get(), v_lugar.get(), v_tarea.get(),v_salario.get(), v_jornada.get(), v_vacante.get(), v_requisitos.get())
            #self.empresa.eliminar_oferta(oft)
            
            tk.messagebox.showinfo(title='Rechazo', message='Rechazado con exito!')    
        
            ventana.destroy()
        
 
    """
    Parte de la empresa 
    """
    
   
    def empresa_sesion(self):
        
         #Ventana para iniciar sesión
         e_sesion = tk.Toplevel(self.vp)
         #Dimensiones
         e_sesion.geometry("275x150")
         #Título
         e_sesion.title("Apartado para usuarios que buscan un trabajador")    
         
         # Usuario: etiquetas con los espacios para rellenar
         etiqueta_usuario = tk.Label(e_sesion, text= "Usuario:").grid(column=0, row=1)
         usuario2 = tk.StringVar()
         usuario2.set("")
         entry_nombre = tk.Entry(e_sesion, textvariable=usuario2).grid(column=1, row=1)
         etiqueta_contraseña = tk.Label(e_sesion, text= "Contraseña:").grid(column=0, row=2)
         contraseña2 = tk.StringVar()
         contraseña2.set("")
         entry_contraseña = tk.Entry(e_sesion, textvariable=contraseña2).grid(column=1, row=2)
                
         #Comprobar el usuario, indica si no existe. La función empleada también se encuentra en este fichero
         verifica_empresa_parametros = partial(self.usuario_parametros, e_sesion, usuario2)
  
         abrir_registro = partial(self.crear_empresa)
        
         Iniciar=tk.Button(e_sesion,text="Iniciar sesión", font=('Gotham',10), width=20,fg='white',bg='forest green',
                            command=verifica_empresa_parametros).grid(column=1,row=3)
         Cancelar=tk.Button(e_sesion,text="Cancelar",font=('Gotham',10), width=20,fg='white',bg='red',
                                  command= e_sesion.destroy).grid(column=1,row=5)
         Registrarse = tk.Button(e_sesion, text = "Registrarse",font=('Gotham',10), width=20,fg='white',bg='forest green',command = abrir_registro).grid(column=1,row=4)
         # Es para que nuestra ventana este sobre su maestra
         e_sesion.transient()

         #Para evitar interactuar con otra ventana
         e_sesion.grab_set()

         # Wait for the window to end 
         self.vp.wait_window(e_sesion)
        
         if usuario2.get():
            existe = self.dic_easyjob.comprueba_usuario_empresa(usuario2.get(), contraseña2.get())
            print(existe)
            
            """
            Si existe el usuario introducido en el campo, abre un menú exclusivo para la empresa.
            Si no existe, devuelve un mensaje de que no está registrado el usuario

            """
            if existe:
                v_empresa = tk.Toplevel(self.vp)
                # Tiene las siguientes dimensiones
                v_empresa.geometry("400x175")
                # El fondo de la ventana es un verda clarito
                v_empresa.configure(bg= 'pale green')
                v_empresa.title("ESPACIO EMPRESA") 
                
                # Creamos la barra de menu
                barramenu = tk.Menu(v_empresa)
                
                #Items
                MenuEmpresa = tk.Menu(barramenu)
                verdatosaux=partial(self.verdatosempresa, usuario2.get())
                verofertasaux=partial(self.ver_oferta, empresa)
                
                #Comandos
                MenuEmpresa.add_command(label="Añadir oferta", command= self.subir_oferta)
                MenuEmpresa.add_command(label="Ver perfil", command= verdatosaux)
                MenuEmpresa.add_command(label="Salir", command = v_empresa.destroy) 
                
                barramenu.add_cascade(label="Empleado", menu = MenuEmpresa)
                
                #Añadir oferta
                oferta = tk.Button(v_empresa, text = "Añadir oferta", font=('Gotham',20), width=20,fg='blue',
                                         bg='black',command= self.subir_oferta).pack(side=TOP, fill=BOTH, expand=True)
                #Añadir oferta
                ver_oferta = tk.Button(v_empresa, text = "Ver ofertas publicadas", font=('Gotham',20), width=20,fg='blue',
                                         bg='black',command= verofertasaux).pack(side=TOP, fill=BOTH, expand=True)
              
                
                verdatos = tk.Button(v_empresa, text = "Ver perfil",font=('Gotham',20), width=20,fg='blue',
                                        bg='black', command = verdatosaux).pack(side=TOP, fill=BOTH, expand=True )
                
                #Salir
                salir = tk.Button(v_empresa, text = "Exit",font=('Gotham',20), width=20,fg='white',bg='red',
                                         command= v_empresa.destroy).pack(side=TOP, fill=BOTH, expand=True)
                
                v_empresa.config(menu = barramenu)
                v_empresa.mainloop()
                         
                 # Esta ventana este sobre su maestra
                v_empresa.transient()
         
                 #Para evitar interactuar con otra ventana
                v_empresa.grab_set()
         
                 # Wait for the window to end 
                self.vp.wait_window(v_empresa)
            
         else:
                tk.messagebox.showinfo(title='Error', message='No existe el usuario')
                
    
    
    def crear_empresa(self):
        
        #Ventana para poder crearse un perfil
        e_login = tk.Toplevel(self.vp)
        #Dimensiones
        e_login.geometry("375x150")
        #Título
        e_login.title("Añadir perfil")    
        #Etiqueta que indica que tienes que hacer
        etiqueta_incio= tk.Label(e_login, text= "Rellenar con los datos de la empresa:")
        etiqueta_incio.grid(column=0, row=0)
        
        
        # Nombre usuario
        etiqueta_usuario = tk.Label(e_login, text= "Nombre usuario:").grid(column=0, row=1)
        usuario2 = tk.StringVar()
        usuario2.set("")
        entry_usuario = tk.Entry(e_login, textvariable=usuario2).grid(column=1, row=1)
        etiqueta_contraseña = tk.Label(e_login, text= "Contraseña:").grid(column=0, row=2)
        contraseña2 = tk.StringVar()
        contraseña2.set("")
        entry_contraseña = tk.Entry(e_login, textvariable=contraseña2).grid(column=1, row=2)
        
        
        # assignar parcial per a funció, per a poder assignar directament command amb variables
        verifica_usuario_params=partial(self.usuario_parametros, e_login, usuario2)

        # Programar botó
        btnAsignar=tk.Button(e_login,text="Aceptar", #command = e_login.destroy).grid(column=0,row=3)
                             command = verifica_usuario_params).grid(column=0,row=3)
        btnSortir=tk.Button(e_login,text="Cancelar", command = e_login.destroy).grid(column=1,row=3)
        e_login.transient()
        e_login.grab_set()
        self.vp.wait_window(e_login)
        
        
        """
        Comprobar que el usuario no existe
        """
        if usuario2.get():
           existe=self.dic_easyjob.comprueba_usuario_empresa(usuario2.get(), contraseña2.get())
           if not existe:
               e_ingreso = tk.Toplevel(self.vp)
               e_ingreso.geometry("300x350")
               e_ingreso.title("Alta de usuario") 
               
               #Nombre
               etiqueta_nombre = tk.Label(e_ingreso, text= "Nombre:").grid(column=0, row=1)
               e_nombre = tk.StringVar()
               e_nombre.set("")               
               entry_nombre = tk.Entry(e_ingreso, textvariable=e_nombre)
               entry_nombre.grid(column=1, row=1)
               self.etiqueta_mensaje=tk.StringVar()
               e_mens1=tk.Label(e_ingreso, textvariable=self.etiqueta_mensaje).grid(column=2, row=2)

               # Ciudad
               etiqueta_lugar = tk.Label(e_ingreso, text= "Ciudad:").grid(column=0, row=2)
               e_lugar = tk.StringVar()
               e_lugar.set("")
               entry_lugar = tk.Entry(e_ingreso, textvariable=e_lugar)
               entry_lugar.grid(column=1, row=2)
               etiqueta_mensaje=tk.StringVar()
               e_mens2=tk.Label(e_ingreso, textvariable=etiqueta_mensaje).grid(column=2, row=2)

               # Código postal
               etiqueta_cp = tk.Label(e_ingreso, text= "Código postal:").grid(column=0, row=3)
               e_cp = tk.StringVar()
               e_cp.set("")
               entry_cp = tk.Entry(e_ingreso, textvariable=e_cp)
               entry_cp.grid(column=1, row=3)
               etiqueta_mensaje=tk.StringVar()
               e_mens3=tk.Label(e_ingreso, textvariable=etiqueta_mensaje).grid(column=2, row=2)

               # Movil
               
               etiqueta_mv = tk.Label(e_ingreso, text= "Nº de movil").grid(column=0, row=4)
               e_mv = tk.StringVar()
               e_mv.set("")
               entry_mv = tk.Entry(e_ingreso, textvariable=e_mv)
               entry_mv.grid(column=1, row=4)
               etiqueta_mensaje=tk.StringVar()
               e_mens4=tk.Label(e_ingreso, textvariable=etiqueta_mensaje).grid(column=2, row=2)
                      
               # email
               etiqueta_email = tk.Label(e_ingreso, text= "Email:").grid(column=0, row=5)
               e_email = tk.StringVar()
               e_email.set("")
               entry_email = tk.Entry(e_ingreso, textvariable=e_email)
               entry_email.grid(column=1, row=5)
               etiqueta_mensaje=tk.StringVar()
               e_mens5=tk.Label(e_ingreso, textvariable=etiqueta_mensaje).grid(column=2, row=2)
               
               # Nombre usuario
               etiqueta_usuario = tk.Label(e_ingreso, text= "Nombre usuario:",state=tk.DISABLED).grid(column=0, row=6)
               e_usuario = tk.StringVar()
               e_usuario.set(usuario2.get())
               entry_usuario = tk.Entry(e_ingreso, textvariable=e_usuario, state=tk.DISABLED).grid(column=1, row=6)
               
               #Contraseña
               etiqueta_contraseña = tk.Label(e_ingreso, text= "Contraseña:", state = tk.DISABLED).grid(column=0, row=7)
               e_contraseña = tk.StringVar()
               e_contraseña.set(contraseña2.get())
               entry_contraseña = tk.Entry(e_ingreso, textvariable=e_contraseña, state = tk.DISABLED)
               entry_contraseña.grid(column=1, row=7)
              
               
               
               crear_empresa_param=partial(self.verificar_entrada,e_ingreso, e_nombre, e_lugar, e_cp, e_mv, e_email, e_usuario, e_contraseña)
                
                
                
        
                # Botón para guardar y una para salir (todavía no sabemos como guardar la función añadida, por ahora ambos botones son para salir)
               Aceptar=tk.Button(e_ingreso,text="Aceptar", font=('Gotham',10), width=20,fg='white',bg='forest green',
                                       command= crear_empresa_param).grid(column=1,row=8)
               Cancelar=tk.Button(e_ingreso,text="Salir",font=('Gotham',10), width=20,fg='white',bg='red',
                                         command= e_ingreso.destroy).grid(column=1,row=9)
        
        
                # Esta ventana este sobre su maestra
               e_ingreso.transient()
        
                #Para evitar interactuar con otra ventana
               e_ingreso.grab_set()
        
                # Wait for the window to end 
               self.vp.wait_window(e_ingreso)
          
          
           else: 
                ventana_error=tk.Toplevel(self.vp)
                ventana_error.geometry("300x300")
                ventana_error.title("IR A INICIAR SESIÓN")
                etiqueta_error = tk.Label(ventana_error, text= "Ya existe una cuenta, por favor ir a iniciar sesion para acceder a ella").grid(column=1, row=1)
                Cancelar=tk.Button(ventana_error,text="Salir",font=('Gotham',10), width=20,fg='white',bg='red',
                                          command= ventana_error.destroy).grid(column=1,row=2)
         
                self.vp.wait_window(ventana_error)  
                

        
    def verificar_entrada(self,e_ingreso, e_nombre, e_lugar, e_cp, e_mv, e_email, e_usuario, e_contraseña):
    # Verificar que los campos requeridos no estén vacíos
        if e_nombre.get() == "" or e_lugar.get() == "" or e_cp.get() == "" or e_mv.get() == "" or e_email.get() == "" or e_contraseña.get() == "":
           tk.messagebox.showinfo(title='Error', message='Falta por rellenar algún hueco')
        else:
            self.dic_easyjob.alta_empresa(e_nombre.get(), e_lugar.get(), e_cp.get(), e_mv.get(), e_email.get(), e_usuario.get(), e_contraseña.get())
            tk.messagebox.showinfo(title='Añadido', message='Añadido el usuario!')
            e_ingreso.destroy()
    
    # Cerrar la ventana
        e_ingreso.destroy()
        
                
                

     #funcion que abre una ventana con todos las datos del usuario          
    def verdatosempresa (self, usuario):
        
       existe = self.dic_easyjob.muestra_empresa(usuario)
       voo = tk.Toplevel(self.vp)
       voo.geometry("300x350")
       voo.title("Alta de usuario") 
                
       # Nombre usuario
       etiqueta_usuario = tk.Label(voo, text= "Usuario:").grid(column=0, row=6)
       v_usuario = tk.StringVar()
       v_usuario.set(existe.usuario)
       entry_usuario = tk.Entry(voo, textvariable=v_usuario, state=tk.DISABLED).grid(column=1, row=6)

       # Nombre
       etiqueta_nombre = tk.Label(voo, text= "Nombre:").grid(column=0, row=1)
       v_nombre = tk.StringVar()
       v_nombre.set(existe.nombre)
       entry_nombre = tk.Entry(voo, textvariable=v_nombre,state=tk.DISABLED).grid(column=1, row=1)
         
        # Ciudad

       etiqueta_ciudad = tk.Label(voo, text= "Ciudad:").grid(column=0, row=2)
       eciudad = tk.StringVar()
       eciudad.set(existe.lugar)
       entry_apellidos = tk.Entry(voo, textvariable=eciudad,state=tk.DISABLED).grid(column=1, row=2)
           
       # Codigo postal
       etiqueta_cp = tk.Label(voo, text= "Codigo postal").grid(column=0, row=3)
       ecp = tk.StringVar()
       ecp.set(existe.codigop)
       entry_cp = tk.Entry(voo, textvariable=ecp, state=tk.DISABLED).grid(column=1, row=3)
         
       # Movil
       etiqueta_mv = tk.Label(voo, text= "Movil").grid(column=0, row=4)
       emv = tk.StringVar()
       emv.set(existe.movil)
       e_mv = tk.Entry(voo, textvariable=emv,state=tk.DISABLED).grid(column=1, row=4)
         
       # Email
       etiqueta_mail = tk.Label(voo, text= "Email:").grid(column=0, row=5)
       email = tk.StringVar()
       email.set(existe.email)
       e_mail = tk.Entry(voo, textvariable=email,state=tk.DISABLED).grid(column=1, row=5)
             
                
       # Contraseña
       etiqueta_pass = tk.Label(voo, text= "Contraseña:").grid(column=0, row=7)
       epass = tk.StringVar()
       epass.set(existe.contraseña)
       entry_pass = tk.Entry(voo, textvariable=epass,state=tk.DISABLED).grid(column=1, row=7)

       exitboton=Button(voo,text="Salir",font=('Gotham',10), width=20,fg='white',bg='red',
                                          command= voo.destroy).grid(column=1,row=12)
         
         
       # Esta ventana este sobre su maestra
       voo.transient()
         
       #Para evitar interactuar con otra ventana
       voo.grab_set()
         
       # Wait for the window to end 
       self.vp.wait_window(voo)
       
    def subir_oferta(self):
        
        #Ventana para poder subir una oferta
        v_oferta = tk.Toplevel(self.vp)
        #Dimensiones de la ventana
        v_oferta.geometry("600x350")
        v_oferta.title("Añadir oferta")    
        #Etiqueta que indica que tienes que hacer
        etiqueta_intro= tk.Label(v_oferta, text= "Rellena la información para tu anuncio:")
        etiqueta_intro.grid(column=0, row=0)

        # Titulo
        etiqueta_titulo = tk.Label(v_oferta, text= "Oferta:").grid(column=0, row=1)
        v_titulo = tk.StringVar()
        v_titulo.set("")
        entry_titulo = tk.Entry(v_oferta, textvariable=v_titulo)
        entry_titulo.grid(column=1, row=1)
        
        # Lugar:
        etiqueta_lugar = tk.Label(v_oferta, text= "Lugar:").grid(column=0, row=2)
        v_lugar = tk.StringVar()
        v_lugar.set("")
        entry_lugar = tk.Entry(v_oferta, textvariable=v_lugar)
        entry_lugar.grid(column=1, row=2)
  
        # Tarea/posición 
        etiqueta_tarea = tk.Label(v_oferta, text= "Tarea a realizar:").grid(column=0, row=3)
        v_tarea = tk.StringVar()
        v_tarea.set("")
        entry_tarea = tk.Entry(v_oferta, textvariable=v_tarea)
        entry_tarea.grid(column=1, row=3)

        # Salario
        etiqueta_salario = tk.Label(v_oferta, text= "Salario:").grid(column=0, row=4)
        v_salario = tk.StringVar()
        v_salario.set("")
        entry_salario = tk.Entry(v_oferta, textvariable=v_salario)
        entry_salario.grid(column=1, row=4)

        # Jornada
        #etiqueta_jornada = tk.Label(v_oferta, text= "Jornada:").grid(column=0, row=5)
        #v_jornada = tk.Spinbox(v_oferta, values=('Jornada completa diurna','Jornada completa nocturna','Jornada parcial (mañanas)','Jornada parcial (tardes)'))
        #v_jornada.grid(column=1, row=5)
        etiqueta_jornada = tk.Label(v_oferta, text= "Jornada:").grid(column=0, row=5)
        v_jornada = tk.StringVar()
        v_jornada.set("")
        entry_jornada = tk.Entry(v_oferta, textvariable=v_jornada)
        entry_jornada.grid(column=1, row=5)        
        # Vacantes
        etiqueta_vacante = tk.Label(v_oferta, text= "Número de vacantes:").grid(column=0, row=6)
        v_vacante = tk.StringVar()
        v_vacante.set("")
        entry_vacante = tk.Entry(v_oferta, textvariable=v_vacante)
        entry_vacante.grid(column=1, row=6)
        
        # Requisitos
        etiqueta_requistos = tk.Label(v_oferta, text= "Introduce los requisios básicos que esperas en un trabajador:").grid(column=0, row=7)
        v_requisitos = tk.StringVar()
        v_requisitos.set("")
        entry_requisitos = tk.Entry(v_oferta, textvariable=v_requisitos)
        entry_requisitos.grid(column=1, row=7)
        
        
        crear_oferta_param=partial(self.crear_oferta_aux, v_oferta, v_titulo, v_lugar, v_tarea, v_salario, v_jornada,v_vacante, v_requisitos)
         
         
         
        # Botón para publicar y otro para salir (todavía no sabemos como guardar la función añadida, por ahora ambos botones son para salir)
        Publicar=tk.Button(v_oferta,text="Publicar", font=('Gotham',10), width=20,fg='white',bg='forest green',
                                 command= crear_oferta_param).grid(column=1,row=8)
        Cancelar=tk.Button(v_oferta,text="Salir",font=('Gotham',10), width=20,fg='white',bg='red',
                                 command= v_oferta.destroy).grid(column=1,row=9)
        
        # Esta ventana este sobre su maestra
        v_oferta.transient()

        #Para evitar interactuar con otra ventana
        v_oferta.grab_set()

        # Wait for the window to end 
        self.vp.wait_window(v_oferta)
  

        v_oferta.mainloop()
        
 
    def crear_oferta_aux(self, v_oferta, v_titulo, v_lugar, v_tarea, v_salario, v_jornada,v_vacante, v_requisitos):
        #if v_titulo.get() == "" or v_lugar.get() == "" or v_tarea.get() == "" or v_salario.get() == "" or v_jornada.get() == "" or v_requisitos.get() == "":
           # tk.messagebox.showinfo(title='Error', message='Falta por rellenar algún hueco')
        if not all([v_titulo.get(), v_lugar.get(),v_tarea.get(),v_salario.get(),v_jornada.get(),v_vacante.get(),v_requisitos.get()]):
            tk.messagebox.showinfo(title='Error', message='Alguno de los campos está vacío')
        else: 
            #self.guardar_oferta(v_titulo, v_lugar, v_tarea,v_salario, v_jornada,v_vacante, v_requisitos)
            oft = oferta(v_titulo.get(), v_lugar.get(), v_tarea.get(),v_salario.get(), v_jornada.get(), v_vacante.get(), v_requisitos.get())
            self.empresa.agregar_oferta(oft)
            
            tk.messagebox.showinfo(title='Añadida', message='Oferta añadida!')    
        
            v_oferta.destroy()
   
    """
    ver_oferta()
    """
          
    def ver_oferta(self, empresa):
        ofertas = self.empresa.ver_ofertas()
        see_oferta = tk.Toplevel(self.vp)
        see_oferta.geometry("600x350")
        see_oferta.title("Ofertas disponibles")
        
        # Encabezado de la tabla
        tk.Label(see_oferta, text="Nombre").grid(row=0, column=0)
        tk.Label(see_oferta, text="Lugar").grid(row=0, column=1)
        tk.Label(see_oferta, text="Tarea").grid(row=0, column=2)
        tk.Label(see_oferta, text="Salario").grid(row=0, column=3)
        tk.Label(see_oferta, text="Jornada").grid(row=0, column=4)
        tk.Label(see_oferta, text="Vacantes").grid(row=0, column=5)
        tk.Label(see_oferta, text="Requisitos").grid(row=0, column=6)
    
        # Iterar a través de todas las ofertas y mostrarlas en la tabla
        i = 1
        for oferta in ofertas:
            nombre, lugar, tarea, salario, jornada, vacantes, requisitos = oferta
                    
            tk.Label(see_oferta, text=nombre).grid(column=0, row=i)
            tk.Label(see_oferta, text=lugar).grid(column=1, row=i)
            tk.Label(see_oferta, text=tarea).grid(column=2, row=i)
            tk.Label(see_oferta, text=salario).grid(column=3, row=i)
            tk.Label(see_oferta, text=jornada).grid(column=4, row=i)
            tk.Label(see_oferta, text=vacantes).grid(column=5, row=i)
            tk.Label(see_oferta, text=requisitos).grid(column=6, row = i)

            i = i + 1

        # Botón para salir de la ventana
        Cancelar = tk.Button(see_oferta, text="Salir", font=('Gotham',10), width=20, fg='white', bg='red', command=see_oferta.destroy)
        Cancelar.grid(row=10, column=7)
    
        # Hacer que la ventana esté sobre su maestra
        see_oferta.transient()
    
        # Para evitar interactuar con otra ventana
        see_oferta.grab_set()
    
        # Esperar a que la ventana se cierre
        self.vp.wait_window(see_oferta)


        

    def oferta_parametros(self, ventana, titulo):
        """
        Auxiliar function to be able to send messageboxes
        """ 
        # Mirar si algun esta empty
        if not all([titulo.get()]):
            tk.messagebox.showinfo(title='Error', message='El campo está vacío')
        else:
            ventana.destroy()
            
```
